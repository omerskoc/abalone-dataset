```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
np.random.seed(0)
```

```python
df=pd.read_csv("abalone.csv")
df
#check data has been read in properly
```

```python
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
##checking correlation 
```

```python
y = df["Rings"] 
df.drop('Rings', axis = 1, inplace = True)
df
##Set the target value and drop from data 
```

```python
df = pd.get_dummies(df)
#### fixing dummies
```

```python
df.drop('Sex_M', axis = 1, inplace = True)
df
x=df
###droping extra dummy variable
# Define the feature set X.
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = .30, random_state = 33)
X_train
###train and test split
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler() 
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train= pd.DataFrame(X_train, columns=["Length","Diameter","Height", "Whole weight","Shucked weight","Viscera weight","Shell weight","Sex_F","Sex_I"])
X_test = pd.DataFrame(X_test, columns=["Length","Diameter","Height", "Whole weight","Shucked weight","Viscera weight","Shell weight","Sex_F","Sex_I"])
X_train
# Standardization
```

```python
from mlxtend.feature_selection import ExhaustiveFeatureSelector as EFS
# Create lienar Regression classifier
lmodel = LinearRegression()
efs = EFS(lmodel, 
           min_features=1,
           max_features=9,
           scoring='neg_mean_squared_error',
           print_progress=True,
           cv=5)

feature_names = ("Length","Diameter","Height", "Whole weight","Shucked weight","Viscera weight","Shell weight","Sex_F","Sex_I")
efs = efs.fit(X_train, y_train, custom_feature_names=feature_names)

print('Best neg_mean_squared_error score: %.2f' % efs.best_score_)
print('Best subset (indices):', efs.best_idx_)
print('Best subset (corresponding names):', efs.best_feature_names_)
```

```python
pd.DataFrame.from_dict(efs.get_metric_dict()).T.sort_values('avg_score',ascending=False)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
lambdas = np.logspace(-4,-0.5,60)
scoresCV = []
for l in lambdas:
    RidgeReg = Ridge(alpha=l)
    RidgeReg.fit(X_train, y_train)    
    scoreCV = cross_val_score(RidgeReg, X_train, y_train, scoring='neg_mean_squared_error',
                             cv=KFold(n_splits=10, shuffle=True,
                                            random_state=1))
    scoresCV.append([l,-1*np.mean(scoreCV)])
df = pd.DataFrame(scoresCV,columns=['Lambda','Validation Error'])
df
```

```python
df.sort_values(by=['Validation Error'])
```
